```python
h(x) = x^3 - 2*x^2 - 5*x + 6
p = plot(h(x), x, -5, 5, ymin = -10, ymax = 10)
p
```

```python
cps = solve(diff(h(x),x) == 0, x)
cps
```

```python
cp0 = cps[0].rhs().n()
cp0
```

```python
cp1 = cps[1].rhs().n()
cp1
```

```python
p = p + point((cp0, h(cp0)), color = "red", pointsize = 100) 
p = p + point((cp1, h(cp1)), color = "red", pointsize = 100) 
## size also works for pointsize
p
## goals: tangent lines, point coordinate labels, inflection point
```
