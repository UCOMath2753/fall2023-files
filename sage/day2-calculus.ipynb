```python
f(x) = x^2 + exp(-x) + 1/x^3
f(x)
```

```python
plot(f(x), x, -5, 5, ymin = -10, ymax = 10)
```

```python
fp(x) = diff(f(x),x)
fp(x)
```

```python
solve(diff(f(x),x) == 0, x)
```

```python
## solve is limited to algebraic solutions
## here the approach is to use find_root() 
## to get a numerical approxation
```

```python
plot(fp(x), x, -5, 5, ymin = -10, ymax = 10)
```

```python
find_root(fp(x), 1, 2)
## just gives the number, 
## no extra 'formatting'
## just one solution, 
## between lower and upper guesses
```

```python
g(x) = (x+2)*(x-1)*(x-3) + 2
g(x).expand()
```

```python
## goals: find and graph CPs and IPs
## try this and we'll regroup to work 
## through it together
```

```python
p = plot(g(x), x, -5, 5, ymin = -5, ymax = 15, axes_labels = ["$x$", "$y$"])
p
```

```python
## CPs: set g'(x) = 0 and solve
cps = solve(diff(g(x), x) == 0 , x)
cps
```

```python
cp0 = cps[0].rhs().n()
cp1 = cps[1].rhs().n()
print(cp0, cp1)
```

```python
p = p + point((cp0, g(cp0)), size = 50, color = "red", legend_label = "critical point")
p = p + point((cp1, g(cp1)), size = 50, color = "red")
p = p + text((round(cp0,2), round(g(cp0), 2)), (cp0, g(cp0)), vertical_alignment = "bottom")
p
## temptation is to put this in a loop
## teaser for next time
```

```python
## PIPs: set g''(x) = 0 and solve
ips = solve(diff(g(x), x, 2) == 0, x)
ips
```

```python
p = p + point((ips[0].rhs(), g(ips[0].rhs())), size = 50, color = "black", legend_label = "inflection point")
p
```

```python
p.save("calc-graph.png")
```
